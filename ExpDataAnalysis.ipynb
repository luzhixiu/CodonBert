```python
import pandas as pd
import math
```

```python
df=pd.read_csv("expressionPrediction_S288C_full.csv")
```

```python
df["EXP_LOG"]=df["EXP"].apply(math.log)

ax=df["EXP_LOG"].plot.hist(xlabel="EXP_Log")
ax.set_xlabel("S288C Log Expression")
ax.set_xscale('log')
```
