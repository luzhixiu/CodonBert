```python
import findSequenceById as FSBID
import scipy.stats as ss
import CodonLibraries as CL
import CUB_Object
import matplotlib.pyplot as plt
from matplotlib import markers
# get all possible shapes
all_shapes = markers.MarkerStyle.markers.keys()

inputFile="../Data/s288c.fasta"


CUB_Object=CUB_Object.CUBObject(inputFile)
codon_map_to_codonObject=CUB_Object.codon_map_to_codonObject



geneDict = FSBID.findSequenceByID(inputFile)
keyList = []
seqList = []
cnt = 0
for key in geneDict:
    seq = geneDict[key]
    if len(seq) % 3 == 0:
        print(cnt, " sequences are not length of three")
        keyList.append(key)
        seqList.append(seq)
    else:
        cnt += 1

        

        

        



```

```python
import MLE

highPhiSeqList=[]
lowPhiSeqList=[]
phiList=[]

for seq in seqList:
    phiList.append(MLE.calPhiForGene(seq))

percentage=0.1
highCutOff=sorted(phiList)[int(len(phiList)*(1-percentage))]
lowCutoff=sorted(phiList)[int(len(phiList)*(percentage))]

for i in range(len(seqList)):
    phi=phiList[i]
    if phi<lowCutoff:
        lowPhiSeqList.append(seqList[i])
    elif phi>highCutOff:
        highPhiSeqList.append(seqList[i])
    
```

```python
    import pandas as pd
    import seaborn as sns
    
    
    targetAA='LYS'
    rscuLists=[]
    for seq in highPhiSeqList:
        codonList = CL.loadSequence(seq)
        aaList = [codon_map_to_codonObject[codon].aa for codon in codonList]
        rscuList = [codon_map_to_codonObject[codon].RSCU for codon in codonList]
        aaRSCUList=[]
        for i in range(len(aaList)):
            aa=aaList[i]
            if aa==targetAA:
                aaRSCUList.append(rscuList[i])
        rscuLists.append(aaRSCUList[:10])
    print(set(aaList))

        
```

```python
import plotBST as pb
tarzan=pb.createBSTfromMultipleRSCUList(rscuLists)
dot=pb.visualize_tree(tarzan,outputFile="high_glu")
```
